### Q1 â€” Exploratory vs. Explanatory
**Prompt:** *"Explain the difference between exploratory and explanatory visualization."*
**Hints:** Consider audience, purpose, level of polish, and how much iteration you expect.

ðŸ“ **Your response:**
**Exploratory Visualization:**
*   **Audience:** Typically the analyst or data scientist themselves.
*   **Purpose:** To understand the data, find patterns, identify outliers, and generate hypotheses. It's about discovery.
*   **Level of polish:** Often rough and unrefined. Speed and flexibility are more important than aesthetics.
*   **Iteration:** Highly iterative. You might create many plots quickly, discarding most, to see what insights emerge.

**Explanatory Visualization:**
*   **Audience:** A specific audience (e.g., executives, clients, the public) who need to understand a particular finding or story from the data.
*   **Purpose:** To communicate a clear message, tell a story, and support an argument or conclusion. It's about communication.
*   **Level of polish:** Highly polished and refined. Design principles are crucial for clarity and impact.
*   **Iteration:** Less iterative in the creation of the final visual, but may involve iteration based on feedback from the target audience to improve clarity.

### Q2 â€” Chart Types & When to Use Them
**Prompt:** *"Describe 5 common chart types (e.g., bar, line, scatter, histogram, map) and the type of data analysis question each is best suited to answer."*  

For each chart, list: **Question archetype**, **Data shape**, **Pitfalls**, **Alternatives**.

ðŸ“ **Your response:**
1.  **Bar:**
    *   Question archetype: Comparison of discrete categories. "How does X compare to Y?" or "What's the distribution across categories?"
    *   Data shape: One categorical variable and one quantitative variable.
    *   Pitfalls: Using too many categories (becomes unreadable), not starting the y-axis at zero (misleading scale), using 3D effects.
    *   Alternatives: Column chart (vertical bars), stacked bar chart (for parts of a whole), Lollipop chart (for sparse data).
2.  **Line:**
    *   Question archetype: Trends over time or sequential data. "How has X changed over time?" or "What is the relationship between two continuous variables where one is sequential?"
    *   Data shape: One sequential/time-based variable and one or more quantitative variables.
    *   Pitfalls: Too many lines (clutter), choosing an inappropriate time scale, using lines for non-sequential data.
    *   Alternatives: Area chart, scatter plot (if the sequential aspect isn't primary).
3.  **Scatter:**
    *   Question archetype: Relationship or correlation between two quantitative variables. "Is there a relationship between X and Y?" or "Are there clusters of data points?"
    *   Data shape: Two quantitative variables. Can include a third quantitative variable or a categorical variable via point size or color.
    *   Pitfalls: Overplotting (too many points), misinterpreting correlation as causation, using for categorical data.
    *   Alternatives: Bubble chart (with size as a third variable), hexbin plot (for dense data).
4.  **Histogram:**
    *   Question archetype: Distribution of a single quantitative variable. "How is the data for variable X distributed?" or "What is the frequency of values within different ranges?"
    *   Data shape: One quantitative variable.
    *   Pitfalls: Choosing inappropriate bin sizes (can hide or exaggerate features), misinterpreting as a bar chart (bins represent ranges, not categories).
    *   Alternatives: Density plot, box plot (shows key summary statistics of the distribution).
5.  **Map:**
    *   Question archetype: Geographic distribution or patterns. "Where are the highest/lowest values located?" or "How does X vary across regions?"
    *   Data shape: Geographic data (latitude/longitude, country, state, zip code, etc.) and one or more quantitative or categorical variables.
    *   Pitfalls: Choosing a misleading projection, using choropleth maps with unequal population densities, difficulty comparing non-adjacent regions.
    *   Alternatives: Symbol map (points on a map), network map (for connections between locations).

### Create a Synthetic Airline Dataset (if you don't have one)
Use this dataset to prototype visualizations and dashboards. It contains realistic columns for the KPIs you defined. You can always swap in your BigQuery data later.


````python
# Generate a synthetic airline dataset
import pandas as pd
import numpy as np
from datetime import datetime, timedelta

np.random.seed(467)

dates = pd.date_range(start='2024-01-01', end='2024-06-30', freq='D')
carriers = ['AA', 'DL', 'UA', 'WN', 'AS']
origins = ['ATL','DFW','DEN','ORD','LAX','SEA','SFO','PHX','CLT','MCO']
dests =   ['JFK','LGA','BOS','SFO','LAX','SEA','MIA','IAH','IAD','PHL']

rows = []
for d in dates:
    for carrier in carriers:
        # number of flights per day per carrier
        n = np.random.randint(80, 140)
        for _ in range(n):
            o = np.random.choice(origins)
            de = np.random.choice(dests)
            if o == de:
                continue
            distance = np.random.randint(300, 2800)
            seats = np.random.choice([90,120,150,180,210])
            load_factor = np.clip(np.random.normal(0.82, 0.07), 0.55, 0.98)
            pax = int(seats * load_factor)
            dep_delay = int(np.random.normal(8, 15))
            arr_delay = dep_delay + int(np.random.normal(0, 12))
            dep_delay = max(dep_delay, -5)
            arr_delay = max(arr_delay, -10)
            cancelled = np.random.rand() < 0.025
            diverted = np.random.rand() < 0.005
            revenue = pax * np.random.uniform(120, 420)
            casm = np.random.uniform(0.08, 0.16)  # cost per available seat mile
            cost = casm * seats * distance
            fuel_burn = distance * np.random.uniform(3.0, 4.5)  # gallons per seat-mile proxy
            nps = np.random.normal(35, 15)  # if you want a customer metric proxy
            rows.append({
                'date': d,
                'carrier': carrier,
                'origin': o,
                'dest': de,
                'distance_miles': distance,
                'seats': seats,
                'passengers': pax,
                'dep_delay_min': dep_delay,
                'arr_delay_min': arr_delay,
                'cancelled': int(cancelled),
                'diverted': int(diverted),
                'revenue_usd': round(revenue, 2),
                'cost_usd': round(cost, 2),
                'fuel_burn_gal': round(fuel_burn, 1),
                'nps': round(nps, 1)
            })

df_flights = pd.DataFrame(rows)

# Derived KPIs
df_flights['on_time'] = (df_flights['arr_delay_min'] <= 15).astype(int)
df_flights['load_factor'] = (df_flights['passengers'] / df_flights['seats']).round(3)
df_flights['yield_usd_per_pax_mile'] = (df_flights['revenue_usd'] / (df_flights['passengers'] * df_flights['distance_miles']).clip(lower=1)).round(4)
df_flights['asm'] = df_flights['seats'] * df_flights['distance_miles']  # available seat miles
df_flights['casm'] = (df_flights['cost_usd'] / df_flights['asm']).round(4)
df_flights['month'] = df_flights['date'].dt.to_period('M').astype(str)

df_flights.head()
````

````python
# Compute On-Time Rate by carrier
on_time_by_carrier = df_flights.groupby('carrier')['on_time'].mean().sort_values(ascending=False)

# Plotting
plt.figure(figsize=(10, 6))
sns.barplot(x=on_time_by_carrier.index, y=on_time_by_carrier.values, hue=on_time_by_carrier.index, palette='viridis', legend=False)

# Add percentage labels on bars
for index, value in enumerate(on_time_by_carrier.values):
    plt.text(index, value + 0.01, f'{value:.1%}', ha='center')

plt.title('On-Time Rate by Carrier')
plt.xlabel('Carrier')
plt.ylabel('On-Time Rate')
plt.ylim(0, 1) # Set y-axis limit from 0 to 1 for percentage
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
````

````python
# Plotting Cost vs. Revenue with CASM coloring
plt.figure(figsize=(10, 8))
scatter_plot = sns.scatterplot(data=df_flights, x='cost_usd', y='revenue_usd', hue='casm', palette='viridis', s=20, alpha=0.6)

# Add a diagonal y=x reference line
max_value = max(df_flights['cost_usd'].max(), df_flights['revenue_usd'].max())
plt.plot([0, max_value], [0, max_value], color='red', linestyle='--', label='Cost = Revenue')

plt.title('Cost vs. Revenue by CASM')
plt.xlabel('Cost (USD)')
plt.ylabel('Revenue (USD)')
plt.legend(title='CASM')
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
````

````python
# Compute mean load factor by route
load_factor_by_route = df_flights.groupby(['origin', 'dest'])['load_factor'].mean().reset_index()
load_factor_by_route['route'] = load_factor_by_route['origin'] + ' - ' + load_factor_by_route['dest']

# Take the top 15 routes by average load factor
top_15_routes = load_factor_by_route.sort_values(by='load_factor', ascending=False).head(15)

# Plotting
plt.figure(figsize=(10, 8))
sns.barplot(x='load_factor', y='route', data=top_15_routes, hue='route', palette='viridis', legend=False)

plt.title('Top 15 Routes by Average Load Factor')
plt.xlabel('Average Load Factor')
plt.ylabel('Route')
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
````

````python
# Plotting a histogram of arrival delays
plt.figure(figsize=(10, 6))
sns.histplot(df_flights['arr_delay_min'], bins=50, kde=True, color='skyblue')

# Add a vertical reference line at 0
plt.axvline(0, color='red', linestyle='--', label='0 Minutes Delay')

# Annotate percent of flights within [-5, 15] minutes
on_time_flights = df_flights[(df_flights['arr_delay_min'] >= -5) & (df_flights['arr_delay_min'] <= 15)]
on_time_percent = len(on_time_flights) / len(df_flights) * 100

plt.text(15, plt.ylim()[1]*0.9, f'{on_time_percent:.1f}% flights within [-5, 15] min', ha='left', color='darkgreen')


plt.title('Distribution of Arrival Delays')
plt.xlabel('Arrival Delay (minutes)')
plt.ylabel('Frequency')
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
````

````python
# Aggregate to daily average arrival delay
daily_avg_delay = df_flights.groupby('date')['arr_delay_min'].mean().reset_index()

# Calculate 7-day rolling average
daily_avg_delay['rolling_avg_delay'] = daily_avg_delay['arr_delay_min'].rolling(window=7).mean()

# Plotting
plt.figure(figsize=(12, 6))
sns.lineplot(x='date', y='arr_delay_min', data=daily_avg_delay, label='Daily Average Delay', color='skyblue')
sns.lineplot(x='date', y='rolling_avg_delay', data=daily_avg_delay, label='7-Day Rolling Average', color='salmon')

plt.title('Average Arrival Delay over Time')
plt.xlabel('Date')
plt.ylabel('Average Arrival Delay (minutes)')
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.tight_layout()
plt.show()
````
